```python
from skimage.io import imread, imshow
from skimage import img_as_float
import math
%matplotlib inline
```

```python
img = imread('tiger-color.png')
entropy = {}
imshow(img)
```

```python
from matplotlib.pyplot import hist
values1, bin_edges1, patches1 = hist(img.ravel(), bins=range(257))
```

```python
#Энтропия для оригинального изображения
H1 = []
for i in range(256):
    pi = values1[i]/(img.shape[0]*img.shape[1])
    if pi != 0:
        H1.append(-pi * math.log(pi))
    else:
        H1.append(0)
entropy.update({"Picture itself": sum(H1)})
```

```python
def MSE(img, img1): #среднеквадратичная ошибка для каждого полученного изображения
    if len(img.shape) > 2:
        img = img[:,:,0]
    if len(img1.shape) >2:
        img1 = img1[:,:,0]
    mse = 0
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            mse += math.pow((img[i,j] - img1[i,j]), 2)
    mse = mse / (img.shape[1] * img.shape[0])
    return mse
```

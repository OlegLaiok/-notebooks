3.1 Линейная коррекция контраста

```python
from skimage.io import imread, imshow
from skimage import img_as_float
%matplotlib inline
```

```python
img = imread('tiger-low-contrast.png')
imshow(img)
```

```python
from matplotlib.pyplot import hist
```

```python
values, bin_edges, patches = hist(img.ravel(), bins=range(257))
```

```python
#Берется максимальный и минимальный элемент изображения
min = img.min() 
max = img.max()
#По формуле применяется к каждому пикселю
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img[i,j] -= min
        img[i,j] *= 255/(max-min)
```

```python
img = img.astype('uint8')
```

```python
imshow(img)
```

```python
values, bin_edges, patches = hist(img.ravel(), bins=range(257))
```

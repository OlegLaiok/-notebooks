```python
from skimage.io import imread, imshow
from skimage import img_as_float, img_as_ubyte, img_as_uint
from scipy.signal import convolve2d
%matplotlib inline
import numpy as np
```

```python
img = imread('tiger-gray-small.png')
imshow(img)
```

```python
kernel = np.array([[-1, -2, -1], [-2, 22, -2], [-1, -2, -1]]) # не забыть умноижить на 1/10
kernel = kernel
def sharp_filter(img, kernel):
    #if img.dtype != "float64":
    #    img = img_as_float(img.copy())
    #else:
    #    img = img.copy()
    img = img.copy()
    box_size = kernel.shape[0]
    filtered_img = np.zeros((img.shape[0] - box_size + 1, img.shape[1] - box_size + 1))
    for x in range(filtered_img.shape[0]):
        for y in range(filtered_img.shape[1]):
            filtered_img[x][y] = sum(sum(img[x:x + box_size, y:y + box_size] * kernel))/10
            print(filtered_img[x][y], x, y)
                
    return filtered_img
```

```python
filtered_img = sharp_filter(img, kernel)
filtered_img = np.clip(filtered_img, 0, 255)
filtered_img = filtered_img.astype('uint8')
imshow(filtered_img)
```

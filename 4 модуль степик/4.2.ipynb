```python
from skimage.io import imread, imshow
from skimage import img_as_float, img_as_ubyte
import numpy as np
%matplotlib inline
```

```python
img = imread('tiger-gray-small.png')
imshow(img)
```

```python
def median_filter(img):
    box_size = 7
    filtered_img = np.zeros((img.shape[0] - box_size + 1, img.shape[1] - box_size + 1))
    for x in range(filtered_img.shape[0]):
        for y in range(filtered_img.shape[1]):
            filtered_img[x][y] = np.median(img[x:x + box_size, y:y + box_size])
    return filtered_img
```

```python
filtered_img = median_filter(img)
imshow(filtered_img.astype('uint8'))
```

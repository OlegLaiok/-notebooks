```python
from skimage.io import imread, imshow, imsave
from skimage import img_as_float, img_as_ubyte, img_as_uint
%matplotlib inline
import numpy as np
```

```python
img = imread('tiger-color.png')
img = img[:,:,0] #Делаем изображение черно-белым

imshow(img)
```

```python
def encode(img, n=1):
    img = img/n
    img = img.astype('uint8')
    return img
```

```python
def decode(img, n=1):  
    img *= n
    img = np.clip(img, 0, 255)
    return img
img = encode(img, 32)
img = decode(img, 32)
imshow(img)
imsave('lab1_1.png', img)
```
